```python
def knn(train, test, k):
    dist = []
    x, y = test
    print(f"({x}, {y})")
    for [a, b, c] in train:
        print(f"({a} - {x})^2 + ({b} - {y})^2 = {(a - x) ** 2 + (b - y) ** 2}")
        dist.append(((a - x) ** 2 + (b - y) ** 2, c))
    dist = sorted(dist, key=lambda x: x[0])
    print(dist[:k])
    vote = {}
    for d, l in dist[:k]:
        vote[l] = vote.get(l, 0) + 1
    print(vote)
    return max(vote.items(), key=lambda x: x[1])[0]

train = [
        (3, 3, 'A'),
        (4, 5, 'B'),
        (2, 2, 'A'),
        (10, 9, 'A'),
        (14, 12, 'B'),
        (17, 19, 'B'),
        (8, 13, 'B'),
]
knn(train, (3, 6), 3)
knn(train, (20, 5), 3)
knn(train, (6, 6), 3)
knn(train, (19, 9), 3)
knn(train, (17, 19), 3)
```
