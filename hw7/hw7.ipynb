```python
1 of the coins is weighted so that it comes up heads with probability 0.1. (Category 1)
2 are weighted so that they come up heads with probability 0.4. (Category 2)
2 are weighted so that they come up heads with probability 0.8. (Category 3)
```

```python
# Q1 randomly pick a coin and flip
import random
import matplotlib.pyplot as plt

coins = [1, 2, 2, 3, 3]
categories = [{'w': [1, 9]},{'w': [4, 6]},{'w': [8, 2]}]
result = []
case = 0
iteration = 0
maxIteration = 10000
minIteration = 10
precision = 0.0001
while maxIteration != iteration:
    coin = random.choice(coins)
    choice = random.choices(['H','T'], categories[coin-1]['w'])
    if choice[0] == 'H':
        case += 1
    result.append(case/(len(result)+1))
    iteration += 1
    if len(result) > minIteration and abs(result[-1] - result[-2]) < precision:
        break
    
print(result[-1])
plt.plot(range(iteration), result)
plt.title('probability')
plt.xlabel('iteration')
plt.ylabel('probability')
plt.show()
```

```python
# Q2 randomly pick a coin and filp and get a tail
# then flip again the same coin and get a tail
import random
import matplotlib.pyplot as plt

coins = [1, 2, 2, 3, 3]
categories = [{'w': [1, 9]},{'w': [4, 6]},{'w': [8, 2]}]
result = []
case = 0
iteration = 0
maxIteration = 10000
minIteration = 10
precision = 0.0001
while maxIteration != iteration:
    coin = random.choice(coins)
    choice = random.choices(['H','T'], categories[coin-1]['w'])
#     if toss a Head, try again. Don't record anything
    if choice[0] == 'H':
        continue

#   if toss a Tail, toss again and record the result
    choice = random.choices(['H','T'], categories[coin-1]['w'])
    if choice[0] == 'T':
        case += 1
    result.append(case/(len(result)+1))
    iteration += 1
    if len(result) > minIteration and abs(result[-1] - result[-2]) < precision:
        break
print("final result is", result[-1])
plt.plot(range(iteration),result)
plt.title('probability')
plt.xlabel('iteration')
plt.ylabel('probability')
plt.show()
```

```python
# Q4 randomly pick a coin to flip and get tail
# then pick another coin from the remaining coins and flip and get a tail
import random
import matplotlib.pyplot as plt

coins = [1, 2, 2, 3, 3]
categories = [{'w': [1, 9]},{'w': [4, 6]},{'w': [8, 2]}]
result = []
case = 0
iteration = 0
maxIteration = 100000
minIteration = 10
precision = 0.000001
while maxIteration != iteration:
    coin = random.choice(coins)
    choice = random.choices(['H','T'], categories[coin-1]['w'])
#     if toss a Head, try again. Don't record anything
    if choice[0] != 'T':
        continue

#   if toss a Tail, pick from the remaining coins and toss it
#   back up coins before remove a coin
    coins_backup = coins.copy()
    coins.remove(coin)
    coin = random.choice(coins)
    choice = random.choices(['H','T'], categories[coin-1]['w'])
    if choice[0] == 'T':
        case += 1
    result.append(case/(len(result)+1))
    iteration += 1
    if len(result) > minIteration and abs(result[-1] - result[-2]) < precision:
        break
    coins = coins_backup
print("final result is", result[-1])
plt.plot(range(iteration),result)
plt.title('probability')
plt.xlabel('iteration')
plt.ylabel('probability')
plt.show()
```

```python
# Q4 randomly pick a coin and flip and observe
# I choose to switch, and my friend chooses the other side of the first one. (Assume we want to maximize our benefit)
# What is my expectation value
import random
import matplotlib.pyplot as plt

coins = [1, 2, 2, 3, 3]
categories = [{'w': [1, 9]},{'w': [4, 6]},{'w': [8, 2]}]
result = []
money = 0
iteration = 0
maxIteration = 1000000
minIteration = 10
precision = 0.00001
while maxIteration != iteration:
    coin = random.choice(coins)
    choice1 = random.choices(['H','T'], categories[coin-1]['w'])
#     if toss a Head, try again. Don't record anything

#   pick another coin from the remaining coins and toss it
#   back up coins before remove a coin
    coins_backup = coins.copy()
    coins.remove(coin)
    coin = random.choice(coins)
    choice2 = random.choices(['H','T'], categories[coin-1]['w'])
    
    if choice1 != choice2:
        money -= 6
    else:
        money += 7
    result.append(money/(len(result)+1))
    iteration += 1
    if len(result) > minIteration and abs(result[-1] - result[-2]) < precision:
        break
    coins = coins_backup
print("final result is", result[-1])
plt.plot(range(iteration),result)
plt.title('probability')
plt.xlabel('iteration')
plt.ylabel('probability')
plt.show()
```
